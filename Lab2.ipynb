```python
import matplotlib.pyplot as plt  # <-- Make sure this line is included

# Group and sum nitrogen use by county
sum_n_by_county = df.groupby('County')['Total N, 1985 (lb.)'].sum()

# Display the first few results
print(sum_n_by_county.head())

# Plot a histogram of the summed nitrogen values
plt.figure(figsize=(10, 6))
plt.hist(sum_n_by_county, bins=20, edgecolor='black')
plt.title('Histogram of Total Nitrogen Use in 1985 by County')
plt.xlabel('Total N, 1985 (lb.)')
plt.ylabel('Number of Counties')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt  

# Group and sum nitrogen use by county
sum_n_by_county = df.groupby('County')['Total N, 1985 (lb.)'].sum()

# Create a scatter plot
plt.figure(figsize=(12, 6))
plt.scatter(sum_n_by_county.index, sum_n_by_county.values, color='blue')
plt.title('Total Nitrogen Use in 1985 by County')
plt.xlabel('County')
plt.ylabel('Total N, 1985 (lb.)')
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.grid(True)
plt.tight_layout()
plt.show()
# The plot shows the distribution of nitrogen use in the Chesapeake Bay in 1985 by county. The plot shows a relatively normal distribution with outliers on the higher side. Using this, an analyst may conclude that these outliers must be some processing plants.
```
